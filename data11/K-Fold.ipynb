```python
## enumberate 함수: 열거하다
## 많은 양의 데이터인 lsit를 다룰 때 편리하게 쓸 수 있는 기능을 제공
## for-each, for
## for x in list: 리스트에 있는 것은 하나씩 꺼내서 처리하기가 아주 간단
##                index가 없어서 불편!
##                하나씩 꺼내서 출력, 연산하기는 용이
##                입력은 안된다.!
##for i in range(n):리스트의 입력과 출력이 다 됨.
##                index로 list에 있는 값들을 접근해야 하기 때문에
##                코드가 복잡하게 보인다.
```

```python
num_list = [100, 200, 300]
```

```python
for x in num_list:
    print(x+100)
```

```python
for i, x in enumerate(num_list):
    print(i, x)
```

```python
for i, x in enumerate(num_list):
    num_list[i]=x+100
num_list
```
